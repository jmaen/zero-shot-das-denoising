```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
noisy = np.load('noisy.npy')
filtered = np.load('filtered.npy')
median = np.load('median.npy')
normalized = np.load('normalized.npy')
```

```python
clip = 2
fig, ax = plt.subplots(1, 4, figsize=(20, 4))

im = ax[0].imshow(noisy, origin='lower', aspect='auto', cmap='seismic', interpolation='none', vmin=-clip, vmax=clip)
ax[0].set_title('Noisy')
ax[0].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x / 2000}'))
ax[0].yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y / 1000}'))

ax[1].imshow(filtered, origin='lower', aspect='auto', cmap='seismic', interpolation='none', vmin=-clip, vmax=clip)
ax[1].set_title('Band-pass Filter')
ax[1].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x / 2000}'))
ax[1].yaxis.set_major_formatter(plt.FuncFormatter(lambda _, __: ''))

ax[2].imshow(median, origin='lower', aspect='auto', cmap='seismic', interpolation='none', vmin=-clip, vmax=clip)
ax[2].set_title('Median Subtraction')
ax[2].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x / 2000}'))
ax[2].yaxis.set_major_formatter(plt.FuncFormatter(lambda _, __: ''))

ax[3].imshow(normalized, origin='lower', aspect='auto', cmap='seismic', interpolation='none', vmin=-clip, vmax=clip)
ax[3].set_title('Local Normalization')
ax[3].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x / 2000}'))
ax[3].yaxis.set_major_formatter(plt.FuncFormatter(lambda _, __: ''))

fig.subplots_adjust(wspace=0.05)

fig.colorbar(im, ax=ax, pad=0.01)

ax[0].set_ylabel('Distance (km)')
x_center = (ax[0].get_position().x0 + ax[3].get_position().x1) / 2
fig.text(x_center, 0.01, 'Time (s)', ha='center')

plt.savefig('fig_5.3.png', bbox_inches='tight', pad_inches=0.01, dpi=200)
plt.show()
```
