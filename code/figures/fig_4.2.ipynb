```python
import numpy as np
import matplotlib.pyplot as plt
from colors import *
```

```python
celeba_psnrs = np.load('celeba_psnrs.npy')
celeba_sps = np.load('celeba_sps.npy')
cbsd68_psnrs = np.load('cbsd68_psnrs.npy')
cbsd68_sps = np.load('cbsd68_sps.npy')
```

```python
celeba_sps
```

```python
cbsd68_sps
```

```python
celeba_mean = celeba_psnrs.mean(axis=0)
celeba_std = celeba_psnrs.std(axis=0)
celeba_sp = celeba_sps.mean()

cbsd68_mean = cbsd68_psnrs.mean(axis=0)
cbsd68_std = cbsd68_psnrs.std(axis=0)
cbsd68_sp = cbsd68_sps.mean()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 4))

ax[0].plot(celeba_mean, color=BLUE)
ax[0].fill_between(range(1000), celeba_mean - celeba_std, celeba_mean + celeba_std, color=BLUE, alpha=0.3)
ax[0].axvline(celeba_sp, linestyle='dashed', color=RED)
ax[0].set_title(u'CelebA (20\N{thin space}dB)')

ax[1].plot(cbsd68_mean, color=BLUE)
ax[1].fill_between(range(1000), cbsd68_mean - cbsd68_std, cbsd68_mean + cbsd68_std, color=BLUE, alpha=0.3)
ax[1].axvline(cbsd68_sp, linestyle='dashed', color=RED)
ax[1].set_title('CBSD68 (10\N{thin space}dB)')

fig.subplots_adjust(wspace=0.1)

ax[0].set_ylabel('PSNR (dB)')
x_center = (ax[0].get_position().x0 + ax[1].get_position().x1) / 2
fig.text(x_center, 0, 'Iterations', ha='center')

plt.savefig('fig_4.2.png', bbox_inches='tight', pad_inches=0.01, dpi=200)
plt.show()
```
