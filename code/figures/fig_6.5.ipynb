```python
import numpy as np
import matplotlib.pyplot as plt
from colors import *
```

```python
noisy = np.load('strong_noisy.npy')
with_norm = np.load('strong_sg.npy')
without_norm = np.load('strong_sg_without.npy')
```

```python
clip = 2
fig, ax = plt.subplots(1, 3, figsize=(20, 4))

im = ax[0].imshow(noisy, origin='lower', aspect='auto', cmap='seismic', interpolation='none', vmin=-clip, vmax=clip)
ax[0].set_title('Input')
ax[0].xaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y / 1000}'))

ax[1].imshow(with_norm, origin='lower', aspect='auto', cmap='seismic', interpolation='none', vmin=-clip, vmax=clip)
ax[1].set_title('With Local Normalization')
ax[1].xaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y / 1000}'))
ax[1].set_yticklabels([])

ax[2].imshow(without_norm, origin='lower', aspect='auto', cmap='seismic', interpolation='none', vmin=-clip, vmax=clip)
ax[2].set_title('Without Local Normalization')
ax[2].xaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y / 1000}'))
ax[2].set_yticklabels([])

fig.subplots_adjust(wspace=0.1)

fig.colorbar(im, ax=ax, pad=0.02)

ax[0].set_ylabel('Channels')
ax[1].set_xlabel('Time (s)')

plt.savefig('fig_6.5.png', bbox_inches='tight', pad_inches=0.01, dpi=200)
plt.show()
```
