```python
import numpy as np
import matplotlib.pyplot as plt
from colors import *
```

```python
noisy = np.load('strong_noisy.npy')
es = np.load('strong_es.npy')
ddip = np.load('strong_ddip.npy')
tv = np.load('strong_tv.npy')
bm3d = np.load('strong_bm3d.npy')
```

```python
clip = 2
fig, ax = plt.subplots(2, 3, figsize=(18, 5))

im = ax[0][0].imshow(noisy, origin='lower', interpolation='none', cmap='seismic', aspect='auto', vmin=-clip, vmax=clip)
ax[0][1].imshow(es, origin='lower', interpolation='none', cmap='seismic', aspect='auto', vmin=-clip, vmax=clip)
ax[0][2].imshow(ddip, origin='lower', interpolation='none', cmap='seismic', aspect='auto', vmin=-clip, vmax=clip)
ax[1][0].set_visible(False)
ax[1][1].imshow(tv, origin='lower', interpolation='none', cmap='seismic', aspect='auto', vmin=-clip, vmax=clip)
ax[1][2].imshow(bm3d, origin='lower', interpolation='none', cmap='seismic', aspect='auto', vmin=-clip, vmax=clip)

for i in range(2):
    for j in range(3):
        ax[i][j].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'{x/1000:.1f}'))

ax[0][1].set_xticklabels([])
ax[0][2].set_xticklabels([])

ax[0][1].set_yticklabels([])
ax[0][2].set_yticklabels([])
ax[1][2].set_yticklabels([])

ax[0][0].set_ylabel('Channels')
ax[1][1].set_xlabel('Time (s)')

ax[0][0].set_title('Input')
ax[0][1].set_title('DIP (ES)')
ax[0][2].set_title('DDIP')
ax[1][1].set_title('DIP-TV')
ax[1][2].set_title('BM3D')

plt.colorbar(im, ax=[ax[0][0], ax[0][1], ax[1][1], ax[0][2], ax[1][2]], location='right', shrink=1, aspect=40, pad=0.02)

plt.savefig(f'fig_6.3.png', bbox_inches='tight', dpi=200)
plt.show()
```
